```python
import pandas as pd 
import matplotlib.pyplot as plt
import geopandas as gpd
```

```python
geo_ucrania = gpd.read_file("dados/stanford-pp624tm0074-geojson.json")
```

```python
geo_ucrania.head(3)
```

```python
# configurando fontes do mapa
plt.rcParams['font.family'] = 'Times New Roman'

# Figura e eixos
fig, ax = plt.subplots(dpi = 175)

# plotando limites dos estados
geo_ucrania.geometry.plot(edgecolor = 'black', linewidth = .35, ax = ax)

# anotaes
ax.annotate(text = 'Fonte: Universidade do Texas', xy = (.59, .05), xycoords = 'figure fraction', fontsize = 8)

# axis
ax.axis('off')

# titulo
ax.set_title('Limites Administrativos - Ucrnia',
             fontsize = 10,
             loc= "left",
             y = 1.05 
            )

plt.show()
```
